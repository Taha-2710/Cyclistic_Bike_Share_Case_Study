```python
import pandas as pd
import glob
import os
```

```python
def clean_cyclistic(df):
    df["started_at"] = pd.to_datetime(df["started_at"])
    df["ended_at"] = pd.to_datetime(df["ended_at"])
    
    df["ride_duration_sec"] = (
        df["ended_at"] - df["started_at"]
    ).dt.total_seconds()
    
    df["ride_duration_min"] = (df["ride_duration_sec"] / 60).round(2)
    
    df = df[df["ride_duration_sec"] >=60]
    df.loc[:,'Week_Day']=df['started_at'].dt.day_name()
    return df
```

```python
input_file=r"C:\Users\Asus\OneDrive\Desktop\Cyclistic_riding_data_2-01-2026\Uncleaned_trip_data\New folder/*.csv"
output_dir=r"C:\Users\Asus\OneDrive\Desktop\Cyclistic_riding_data_2-01-2026\Cleaned_trip_data"

files=glob.glob(input_file)
for file in files:
    df=pd.read_csv(file) 
    df=clean_cyclistic(df)
    base_name= os.path.basename(file)
    name_without_ext= os.path.splitext(base_name)[0]
    output_file= os.path.join(output_dir,f"{name_without_ext}_cleaned.csv")
    df.to_csv(output_file,index=False)
```
